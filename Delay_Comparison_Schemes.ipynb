```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
data = pd.read_csv("/content/privacy_results.csv")

# Set up plot styles
styles = {
    "PROPOSED": {"marker": "^", "color": "#FFD700", "label": "Proposed Scheme"},
    "BASELINE_CRYPTO_ONLY": {"marker": "s", "color": "#FF8C00", "label": "Baseline Crypto Only"},
    "BASELINE_COOP_SWAP": {"marker": "o", "color": "#1E90FF", "label": "Baseline Coop Swap"}
}

plt.figure(figsize=(6, 4))

# Plot each scheme
for scheme, style in styles.items():
    subset = data[data["Scheme"] == scheme]
    plt.plot(subset["VehicleID"], subset["TotalDelay"],
             linestyle='-', marker=style["marker"], color=style["color"],
             label=style["label"], markersize=5)

# Format axes and legend
plt.xlabel("Vehicle ID", fontsize=10, fontname="Times New Roman")
plt.ylabel("Total Delay (seconds)", fontsize=10, fontname="Times New Roman")
plt.xticks(range(0, 51, 5))
plt.yticks(range(0, 11, 1))
plt.grid(True)
plt.legend(loc="upper center", ncol=3, frameon=False, fontsize=9)
plt.tight_layout()

# Export as vector PDF
plt.savefig("fig6_cleaned.pdf", format="pdf")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
data = pd.read_csv("/content/privacy_results.csv")

# Define styles for each scheme with updated labels
styles = {
    "PROPOSED": {"marker": "^", "color": "#FFD700", "label": "Proposed Scheme"},
    "BASELINE_CRYPTO_ONLY": {"marker": "s", "color": "#FF8C00", "label": "Crypto Only [47]"},
    "BASELINE_COOP_SWAP": {"marker": "o", "color": "#1E90FF", "label": "Coop.Swap [32]"}
}

# Create the figure
plt.figure(figsize=(6, 4))

# Plot each scheme
for scheme, style in styles.items():
    subset = data[data["Scheme"] == scheme]
    plt.plot(subset["VehicleID"], subset["TotalDelay"],
             linestyle='-', marker=style["marker"], color=style["color"],
             label=style["label"], markersize=5)

# Axis formatting
plt.xlabel("Vehicle ID", fontsize=10, fontname="Times New Roman")
plt.ylabel("Total Delay (seconds)", fontsize=10, fontname="Times New Roman")
plt.xticks(range(0, 51, 5))
plt.yticks(range(0, 11, 1))
plt.grid(True)

# Legend formatting (centered above plot)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),
           ncol=3, frameon=False, fontsize=9)

# Adjust layout to fit legend
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Export as vector PDF
plt.savefig("fig6_cleaned.pdf", format="pdf")
```
